Question 1 - What is hypothesis testing in statistics?

Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves testing an assumption (hypothesis) and determining whether there is enough evidence to reject it.

Question 2 - What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:
The null hypothesis (H0) is a statement that there is no effect or no difference.

The alternative hypothesis (H1) is the opposite statement, suggesting that there is an effect or a difference.

Example:
H0: The mean is equal to 50
H1: The mean is not equal to 50

Question 3 - Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It is usually set at 0.05 (5%).

If the p-value is less than α, we reject the null hypothesis; otherwise, we fail to reject it. It controls the risk of making a Type I error.

Question 4 - What are Type I and Type II errors? Give examples of each.

Answer:
Type I error occurs when we reject a true null hypothesis.
Example: Concluding a medicine works when it actually does not.

Type II error occurs when we fail to reject a false null hypothesis.
Example: Concluding a medicine does not work when it actually works.

Question 5 - What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:
A Z-test is used when the population variance is known and the sample size is large (n > 30).

A T-test is used when the population variance is unknown and the sample size is small (n ≤ 30).

The T-test uses the t-distribution, while the Z-test uses the normal distribution

```python
# Question 7 - Hypothesis Testing using Z-test

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Question 8 - Normal Distribution and 95% Confidence Interval

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
std = np.std(data)
n = len(data)

confidence_interval = (mean - 1.96*(std/np.sqrt(n)), mean + 1.96*(std/np.sqrt(n)))

print("Mean:", mean)
print("95% Confidence Interval:", confidence_interval)

plt.hist(data, bins=30)
plt.title("Normal Distribution")
plt.show()
```

```python
# Question 9 - Z-score Calculation and Histogram

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = [(x - mean) / std for x in data]
    return z_scores

data = [10, 20, 30, 40, 50, 60, 70]

z_scores = calculate_z_scores(data)

print("Z-scores:", z_scores)

plt.hist(z_scores, bins=5)
plt.title("Z-score Distribution")
plt.show()

#Z-scores represent how many standard deviations a data point is from the mean. A positive Z-score indicates the value is above the mean, while a negative Z-score indicates it is below the mean.
```
